```python
import pandas as pd

#open the file and get the top 5 rows 
exchange_rates = pd.read_csv("euro-daily-hist_1999_2021.csv")
exchange_rates.head()
```

```python
# get the bottom 5 rows 
exchange_rates.tail()
```

```python
# get the info about the data 

exchange_rates.info()
```

```python
# clean all the column titles 

headers = []
new_headers = []
for col in exchange_rates.columns:
    headers.append(col)
    new_col = col.strip("[]")
    new_headers.append(new_col)

i = 1
while i < len(headers):
    exchange_rates.rename(columns={headers[i]:new_headers[i]},inplace=True)
    i += 1

exchange_rates.rename(columns={headers[0]:'Date'},inplace=True)
    
exchange_rates.head()
```

```python
# change the date column dateDate datatype, sort data in ascending order

exchange_rates['Date'] = pd.to_datetime(exchange_rates['Date'])
exchange_rates.sort_values('Date',inplace=True)
exchange_rates.reset_index(drop=True,inplace=True)
exchange_rates.head()
```

```python
# euro to Canadian cad

euro_to_cad = exchange_rates[['Date','Canadian dollar ']].copy()
euro_to_cad['Canadian dollar '].value_counts()
```

```python
euro_to_cad = euro_to_cad[euro_to_cad['Canadian dollar '] != '-']
euro_to_cad['Canadian dollar '] = euro_to_cad['Canadian dollar '].astype(float)
euro_to_cad.info()
```

```python
# Rolling mean

import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_to_cad['Date'],euro_to_cad['Canadian dollar '])
plt.show()
```

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_cad['Date'],euro_to_cad['Canadian dollar '])
plt.title('Original Values',weight='bold')

for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_cad['Date'],
             euro_to_cad['Canadian dollar '].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')
    
plt.tight_layout() # Auto-adjusts the padding between subplots
plt.show()
```

```python
euro_to_cad['rolling_mean'] = euro_to_cad['Canadian dollar '].rolling(30).mean()
euro_to_cad
```

```python
# EURO - CAD rate over different prime ministers

harper_trudeau = euro_to_cad.copy(
                   )[(euro_to_cad['Date'].dt.year >= 2009) & (euro_to_cad['Date'].dt.year < 2021)]

harper = harper_trudeau[harper_trudeau['Date'].dt.year < 2015]
trudeau = harper_trudeau[(harper_trudeau['Date'].dt.year >= 2015) & (harper_trudeau['Date'].dt.year <= 2021)]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

plt.figure(figsize=(12, 6))
ax1 = plt.subplot(2,2,1)
ax2 = plt.subplot(2,2,2)
ax3 = plt.subplot(2,1,2)
axes = [ax1, ax2, ax3] 

for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.00, 1.25, 1.50, 1.75])
    ax.set_yticklabels(['1.00', '1.25','1.50', '1.75'],
                   alpha=0.3)
    ax.grid(alpha=0.5)   

# Plot haprer

ax1.plot(harper['Date'], harper['rolling_mean'],
        color='blue')
ax1.set_xticklabels(['', '2009', '', '2011', '', '2013', '',
                     '2015'],
                   alpha=0.3)
ax1.text(15100, 1.90, 'Harper', fontsize=18, weight='bold', color='BLUE')
ax1.text(15000, 1.80, '(2009-2015)', weight='bold', alpha=0.3)

 # Plot trudeau 
ax2.plot(trudeau['Date'], trudeau['rolling_mean'],
        color='red')
ax2.set_xticklabels(['', '2015', '', '2017', '', '2019', '',
                     '2021'],
                   alpha=0.3)

ax2.text(17250, 1.90, 'Trudeau', fontsize=18, weight='bold', color='red')
ax2.text(17200, 1.80, '(2015-2021)', weight='bold', alpha=0.3)

# harper - trudea
ax3.plot(harper['Date'], harper['rolling_mean'],
        color='blue')
ax3.plot(trudeau['Date'], trudeau['rolling_mean'],
        color='red')
ax3.grid(alpha=0.5)
ax3.set_xticks([])

ax.text(13700.0, 3.4, 'EURO-CAD rate averaged 1.51 under the last two Canadian prime ministers',fontsize=20, weight='bold')
ax.text(13700.0, 3.3, '''EURO-CAD exchange rates under Steven Harper (2009 - 2015) and Justin Trudeau (2015-2021)''',fontsize=16)

plt.show()
```
